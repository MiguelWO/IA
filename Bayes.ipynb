```python
outlook_dict = {"outlook": ["sunny", "overcast","rain", "total"] , "p":[0,0,0,0],"n":[0,0,0,0],"total":[0,0,0,0]}
temperature_dict = {"temperature":["hot","mild","cool","total"], "p":[0,0,0,0],"n":[0,0,0,0],"total":[0,0,0,0]}
humidity_dict = {"humidity":["high","normal", "total"], "p":[0,0,0],"n":[0,0,0],"total":[0,0,0]}
windy_dict = {"windy":["f","t", "total"], "p":[0,0,0],"n":[0,0,0],"total":[0,0,0]}

# print(outlook_dict)
# print(temperature_dict)
# print(humidity_dict)
# print(windy_dict)


#for ind in data_df.index:
#     outlook = data_df['Outlook'][ind]
#     temperature = data_df['Temperature '][ind]
#     humidity = data_df['Humidity'][ind]
#     windy = data_df['Windy'][ind]
#     class_ = data_df['Class'][ind]
#     # outlook_df["total"][outlook_c.index(outlook)] += 1
#     # temperature_df["total"][temperature_c.index(temperature)] += 1
#     # humidity_df["total"][humidity_c.index(humidity)] += 1
#     # windy_df["total"][windy_c.index(windy)] += 1
#     if class_ == "p":
#         outlook_df["p"][outlook_c.index(outlook)] += 1
#         temperature_df["p"][temperature_c.index(temperature)] += 1
#         humidity_df["p"][humidity_c.index(humidity)] += 1
#         windy_df["p"][windy_c.index(windy)] += 1
#     else:
#         outlook_df["n"][outlook_c.index(outlook)] += 1
#         temperature_df["n"][temperature_c.index(temperature)] += 1
#         humidity_df["n"][humidity_c.index(humidity)] += 1
#         windy_df["n"][windy_c.index(windy)] += 1



# Create a function to count the number of each class

for ind in data_df.index:
    outlook = data_df.iloc[ind, 0]
    temperature = data_df.iloc[ind, 1]
    humidity = data_df.iloc[ind, 2]
    windy = data_df.iloc[ind, 3]
    class_ = data_df.iloc[ind, 4]

    # print(outlook)
    # print(temperature)
    # print(humidity)
    # print(windy)
    # print(class_)

    if class_ == "p":
        outlook_dict["p"][outlook_c.index(outlook)] += 1
        outlook_dict["p"][3] += 1
        outlook_dict["total"][outlook_c.index(outlook)] += 1
        temperature_dict["p"][temperature_c.index(temperature)] += 1
        temperature_dict["p"][3] += 1
        temperature_dict["total"][temperature_c.index(temperature)] += 1
        humidity_dict["p"][humidity_c.index(humidity)] += 1
        humidity_dict["p"][2] += 1
        humidity_dict["total"][humidity_c.index(humidity)] += 1
        windy_dict["p"][windy_c.index(windy)] += 1
        windy_dict["p"][2] += 1
        windy_dict["total"][windy_c.index(windy)] += 1
    else:
        outlook_dict["n"][outlook_c.index(outlook)] += 1
        outlook_dict["n"][3] += 1
        outlook_dict["total"][outlook_c.index(outlook)] += 1
        temperature_dict["n"][temperature_c.index(temperature)] += 1
        temperature_dict["n"][3] += 1
        temperature_dict["total"][temperature_c.index(temperature)] += 1
        humidity_dict["n"][humidity_c.index(humidity)] += 1
        humidity_dict["n"][2] += 1
        humidity_dict["total"][humidity_c.index(humidity)] += 1
        windy_dict["n"][windy_c.index(windy)] += 1
        windy_dict["n"][2] += 1
        windy_dict["total"][windy_c.index(windy)] += 1

outlook_dict["total"][3] = outlook_dict["p"][3] + outlook_dict["n"][3]
temperature_dict["total"][3] = temperature_dict["p"][3] + temperature_dict["n"][3]
humidity_dict["total"][2] = humidity_dict["p"][2] + humidity_dict["n"][2]
windy_dict["total"][2] = windy_dict["p"][2] + windy_dict["n"][2]



# print(outlook_dict)
# print(temperature_dict)
# print(humidity_dict)
# print(windy_dict)

outlook_df = pd.DataFrame(outlook_dict)
temperature_df = pd.DataFrame(temperature_dict)
humidity_df = pd.DataFrame(humidity_dict)
windy_df = pd.DataFrame(windy_dict)


print(outlook_df)
print(temperature_df)
print(humidity_df)
print(windy_df)
```

```python
#Create a function to calculate the probability of each class
outlook_dict_p = {"outlook": ["sunny", "overcast","rain"] ,
                  "p":[outlook_dict["p"][0]/outlook_dict["p"][3],
                       outlook_dict["p"][1]/outlook_dict["p"][3],
                       outlook_dict["p"][2]/outlook_dict["p"][3]],
                  "n":[outlook_dict["n"][0]/outlook_dict["n"][3],
                       outlook_dict["n"][1]/outlook_dict["n"][3],
                       outlook_dict["n"][2]/outlook_dict["n"][3]],
                  "P(x)": [outlook_dict["total"][0]/outlook_dict["total"][3],
                           outlook_dict["total"][1]/outlook_dict["total"][3],
                           outlook_dict["total"][2]/outlook_dict["total"][3]]}

temperature_dict_p = {"temperature":["hot","mild","cool"],
                        "p":[temperature_dict["p"][0]/temperature_dict["p"][3],
                            temperature_dict["p"][1]/temperature_dict["p"][3],
                            temperature_dict["p"][2]/temperature_dict["p"][3]],
                        "n":[temperature_dict["n"][0]/temperature_dict["n"][3],
                                temperature_dict["n"][1]/temperature_dict["n"][3],
                                temperature_dict["n"][2]/temperature_dict["n"][3]],
                        "P(x)": [temperature_dict["total"][0]/temperature_dict["total"][3],
                                temperature_dict["total"][1]/temperature_dict["total"][3],
                                temperature_dict["total"][2]/temperature_dict["total"][3]]}

humidity_dict_p = {"humidity":["high","normal"],
                    "p":[humidity_dict["p"][0]/humidity_dict["p"][2],
                        humidity_dict["p"][1]/humidity_dict["p"][2]],
                    "n":[humidity_dict["n"][0]/humidity_dict["n"][2],
                        humidity_dict["n"][1]/humidity_dict["n"][2]],
                    "P(x)": [humidity_dict["total"][0]/humidity_dict["total"][2],
                            humidity_dict["total"][1]/humidity_dict["total"][2]]}

windy_dict_p = {"windy":["true","false"],
                "p":[windy_dict["p"][0]/windy_dict["p"][2],
                    windy_dict["p"][1]/windy_dict["p"][2]],
                "n":[windy_dict["n"][0]/windy_dict["n"][2],
                    windy_dict["n"][1]/windy_dict["n"][2]],
                "P(x)": [windy_dict["total"][0]/windy_dict["total"][2],
                        windy_dict["total"][1]/windy_dict["total"][2]]}


outlook_df = pd.DataFrame(outlook_dict_p)
temperature_df = pd.DataFrame(temperature_dict_p)
humidity_df = pd.DataFrame(humidity_dict_p)
windy_df = pd.DataFrame(windy_dict_p)

print(outlook_df)
print(temperature_df)
print(humidity_df)
print(windy_df)
```

```python
play = []
no = []
for ind in data_df.index:
    play.append(outlook_dict_p["p"][outlook_c.index(data_df.iloc[ind,0])] *
                temperature_dict_p["p"][temperature_c.index(data_df.iloc[ind,1])] *
                humidity_dict_p["p"][humidity_c.index(data_df.iloc[ind,2])] *
                windy_dict_p["p"][windy_c.index(data_df.iloc[ind,3])]*
                (outlook_dict["p"][3]/outlook_dict["total"][3]))
    no.append(outlook_dict_p["n"][outlook_c.index(data_df.iloc[ind,0])] *
                temperature_dict_p["n"][temperature_c.index(data_df.iloc[ind,1])] *
                humidity_dict_p["n"][humidity_c.index(data_df.iloc[ind,2])] *
                windy_dict_p["n"][windy_c.index(data_df.iloc[ind,3])]*
                (outlook_dict["n"][3]/outlook_dict["total"][3]))

print(play)
print(no)

final = ["yes" if play[i] > no[i] else "no" for i in range(len(play))]
print(final)
```

```python
data_df["Yes"] = play
data_df["No"] = no

data_df["Prediction"] = final


print(data_df)
```
